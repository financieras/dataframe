### Solución alternativa con `groupbay`

El método `df.groupby('City')` en Pandas sirve para agrupar un DataFrame según los valores únicos que haya en la columna "City".

Esto crea un objeto tipo GroupBy que representa los grupos formados, y permite luego aplicar operaciones o funciones sobre cada grupo por separado. Es un proceso de tres pasos: "Split-Apply-Combine":

- **Split:** Se divide el DataFrame en grupos según las ciudades diferentes en la columna "City".
- **Apply:** Se pueden aplicar funciones sobre cada grupo, como calcular la media, suma, conteo, etc.
- **Combine:** Se combinan los resultados en un nuevo DataFrame u otro objeto con el resumen deseado.

Por ejemplo, si quieres saber la cantidad de filas que tiene cada ciudad, usarías:

```python
df.groupby('City').size()
```

Si quisieras la media de otra columna numérica por ciudad, algo como:

```python
df.groupby('City')['otra_columna'].mean()
```

El objeto que devuelve `.groupby()` no es un DataFrame sino un objeto intermedio, que debe usarse junto con funciones de agregación para extraer resultados útiles.

En resumen, `.groupby('City')` es muy útil para segmentar y analizar datos agrupados por ciudades o cualquier otro criterio categórico, facilitando análisis estadísticos y sumarios basados en grupos.

````python
df.groupby('City').size().sort_values(ascending=False)[:10].plot.barh()
````

# Preguntas Task 1.2
¿Cuál es la ciudad entre los 10 ciudades con más anuncios con la media de candidatos más baja?
- Respuesta: Gipuzkoa

````python
# Paso 1: Obtener las 10 ciudades con más anuncios
top10_cities = df['City'].value_counts().index[:10]
print(top10_cities)
print()


# Paso 2: Filtrar DataFrame con solo esas 10 ciudades
df_top10 = df[df['City'].isin(top10_cities)]

# Paso 3: Agrupar por ciudad y calcular la media de candidatos
candidates_mean = df_top10.groupby('City')['Candidates'].mean()
print(candidates_mean)
print()

# Paso 4: Ciudad con la media más baja
city_min_candidates = candidates_mean.idxmin()
min_candidates_value = candidates_mean.min()

print(f"La ciudad entre las 10 con más anuncios con la media de candidatos más baja es {city_min_candidates} con una media de {min_candidates_value:.2f} candidatos.")
````

### Solución del profe

````python
df_res = df.groupby('City')['Candidates'].agg(['count', 'mean']).sort_values('count', ascending = False)
df_res[:10]
````

### Explicación
El patrón “split-apply-combine” en Pandas, que se implementa con `groupby`, es una estrategia muy utilizada para analizar datos agrupados: primero se **divide** (split) el conjunto de datos en grupos según algún criterio, luego se **aplica** (apply) una función a cada grupo por separado, y por último se **combina** (combine) el resultado de cada grupo en una nueva estructura de datos (normalmente un DataFrame).

### Explicación del patrón split-apply-combine en Pandas

- **Split (Dividir):** Se crean grupos a partir de una o más columnas (por ejemplo, todas las filas con la misma ciudad).
- **Apply (Aplicar):** Se ejecuta una función sobre cada grupo. Puede ser una agregación (como contar, calcular la media, suma, etc.), transformación o filtrado.
- **Combine (Combinar):** Los resultados de cada grupo se ensamblan en un nuevo DataFrame o estructura de salida.

Este patrón permite operaciones similares a un “GROUP BY” en SQL para cálculos resumidos o agrupados.[1]

### Explicación de tu línea de código

```python
df_res = df.groupby('City')['Candidates'].agg(['count', 'mean']).sort_values('count', ascending = False)
df_res[:10]
```

- `df.groupby('City')`: Divide el DataFrame en grupos según el valor de la columna ‘City’ (split).
- `['Candidates']`: Selecciona la columna sobre la que se van a aplicar funciones.
- `.agg(['count', 'mean'])`: Aplica las funciones de agregación `count` (cantidad de filas en el grupo) y `mean` (media aritmética de ‘Candidates’) a cada grupo (apply).
- `.sort_values('count', ascending=False)`: Ordena los grupos de mayor a menor según el recuento.
- `df_res[:10]`: Da la tabla resumida para las 10 ciudades con más registros (combine).

### ¿Qué es ‘agg’?

El método `.agg()` en Pandas permite **aplicar una o varias funciones de agregación** (como `count`, `mean`, `sum`, `min`, `max`) sobre los grupos generados por `groupby`. Si pasas una lista, como en `agg(['count', 'mean'])`, Pandas calculará esas funciones para cada grupo y devolverá un DataFrame con los resultados.

### Solución alternativa del profe

````python
df_res = pd.DataFrame(columns=['City', 'Count', 'Cand_aver'])   # Creamos un df vacío con tres columnas

for city in df['City'].value_counts().index[:10]:
    df_city = df[df['City'] == city]
    count = len(df_city)
    mean_cand = df_city['Candidates'].mean()
    df_res.loc[len(df_res.index)] = [city, count, round(mean_cand, 1)]

df_res.sort_values('Cand_aver', ascending=True)
````
