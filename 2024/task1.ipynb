### Solución alternativa con `groupbay`

El método `df.groupby('City')` en Pandas sirve para agrupar un DataFrame según los valores únicos que haya en la columna "City".

Esto crea un objeto tipo GroupBy que representa los grupos formados, y permite luego aplicar operaciones o funciones sobre cada grupo por separado. Es un proceso de tres pasos: "Split-Apply-Combine":

- **Split:** Se divide el DataFrame en grupos según las ciudades diferentes en la columna "City".
- **Apply:** Se pueden aplicar funciones sobre cada grupo, como calcular la media, suma, conteo, etc.
- **Combine:** Se combinan los resultados en un nuevo DataFrame u otro objeto con el resumen deseado.

Por ejemplo, si quieres saber la cantidad de filas que tiene cada ciudad, usarías:

```python
df.groupby('City').size()
```

Si quisieras la media de otra columna numérica por ciudad, algo como:

```python
df.groupby('City')['otra_columna'].mean()
```

El objeto que devuelve `.groupby()` no es un DataFrame sino un objeto intermedio, que debe usarse junto con funciones de agregación para extraer resultados útiles.

En resumen, `.groupby('City')` es muy útil para segmentar y analizar datos agrupados por ciudades o cualquier otro criterio categórico, facilitando análisis estadísticos y sumarios basados en grupos.

````python
df.groupby('City').size().sort_values(ascending=False)[:10].plot.barh()
````

# Preguntas Task 1.2
¿Cuál es la ciudad entre los 10 ciudades con más anuncios con la media de candidatos más baja?
- Respuesta: Gipuzkoa

````python
# Paso 1: Obtener las 10 ciudades con más anuncios
top10_cities = df['City'].value_counts().index[:10]
print(top10_cities)
print()


# Paso 2: Filtrar DataFrame con solo esas 10 ciudades
df_top10 = df[df['City'].isin(top10_cities)]

# Paso 3: Agrupar por ciudad y calcular la media de candidatos
candidates_mean = df_top10.groupby('City')['Candidates'].mean()
print(candidates_mean)
print()

# Paso 4: Ciudad con la media más baja
city_min_candidates = candidates_mean.idxmin()
min_candidates_value = candidates_mean.min()

print(f"La ciudad entre las 10 con más anuncios con la media de candidatos más baja es {city_min_candidates} con una media de {min_candidates_value:.2f} candidatos.")
````
