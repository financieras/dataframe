```python
import pandas as pd
import matplotlib.pyplot as plt
import ast

pth = "bigdata/dataset_reto_modificado.csv"
df = pd.read_csv(pth)


# Convertir strings a listas de Python
skills_matrix = [ast.literal_eval(skill) for skill in df["Skills"]]

print(skills_matrix)
print(skills_matrix[0])
print(skills_matrix[0][0])

# Veces que aparece la skill "python"
contador = 0
for fila in skills_matrix:
    for habilidad in fila:
        if habilidad == "python":
            contador += 1

print(contador)
```

```python
# Calculamos cuantas Skills hay en total
acumulado_habilidades = [elemento for fila in skills_matrix for elemento in fila]
len(acumulado_habilidades)
```

```python
# Vemos cuantos hay de cada clase: "sql server" o "sqlserver"
print(sum(1 for fila in skills_matrix for elemento in fila if elemento == "sqlserver"))
print(sum(1 for fila in skills_matrix for elemento in fila if elemento == "sql server"))
```

```python
from collections import Counter

# Creamos una lista plana con todas las tecnologías
all_technologies = [tech for sublist in skills_matrix for tech in sublist]

# Contamos la frecuencia de cada tecnología
frecuencia_tecnologias = Counter(all_technologies)

# Creamos una lista de tuplas donde el primer elemento es la tecnología y el segundo es su frecuencia
lista_frecuencias = [(tech, freq) for tech, freq in frecuencia_tecnologias.items()]

# Ordenamos la lista de tuplas por la frecuencia en orden descendente
lista_frecuencias.sort(key=lambda x: x[1], reverse=True)

print(lista_frecuencias[:50])
```

```python
import pandas as pd
from collections import defaultdict

# Suponiendo que df es tu dataframe
# Si aún no has importado los datos, primero necesitas cargarlos en un dataframe
# df = pd.read_csv('nombre_del_archivo.csv')

# Crear un diccionario para almacenar la frecuencia de habilidades y el número acumulado de candidatos
skills_freq = defaultdict(int)
candidates_accumulated = defaultdict(int)

# Iterar sobre cada fila del dataframe
for index, row in df.iterrows():
    skills = eval(row['Skills'])  # Convertir la cadena de habilidades en una lista
    candidates = row['Candidates']

    # Actualizar la frecuencia de cada habilidad
    for skill in skills:
        skills_freq[skill] += 1
        candidates_accumulated[skill] += candidates

# Convertir el diccionario en una lista de tuplas y calcular el ratio
result = [(skill, skills_freq[skill], candidates_accumulated[skill], round(candidates_accumulated[skill] / skills_freq[skill], 1)) for skill in skills_freq]

# Seleccionar las 50 tecnologías de mayor frecuencia y ordenarlas por el tercer elemento (ratio) de las tuplas de forma ascendente
result = sorted(result, key=lambda x: x[1], reverse=True)[:50]

# Ordenar la lista por el tercer elemento (ratio) de las tuplas de forma ascendente
result = sorted(result, key=lambda x: x[3])

# Imprimir el resultado
print(result)
```
